```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler,OneHotEncoder,FunctionTransformer
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
```

```python
non_null_cols=data.columns[data.isnull().sum()==0]
print(non_null_cols)
```

```python
no_trans=['MSSubClass', 'MSZoning', 'LotArea', 'Street', 'LotShape',
       'LandContour', 'Utilities', 'LotConfig', 'LandSlope', 'Neighborhood',
       'Condition1', 'Condition2', 'BldgType', 'HouseStyle', 'OverallQual',
       'OverallCond', 'YearBuilt', 'YearRemodAdd', 'RoofStyle', 'RoofMatl',
       'Exterior1st', 'Exterior2nd', 'ExterQual', 'ExterCond', 'Foundation',
       'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF', 'Heating',
       'HeatingQC', 'CentralAir', '1stFlrSF', '2ndFlrSF', 'LowQualFinSF',
       'GrLivArea', 'BsmtFullBath', 'BsmtHalfBath', 'FullBath', 'HalfBath',
       'BedroomAbvGr', 'KitchenAbvGr', 'KitchenQual', 'TotRmsAbvGrd',
       'Functional', 'Fireplaces', 'GarageCars', 'GarageArea', 'PavedDrive',
       'WoodDeckSF', 'OpenPorchSF', 'EnclosedPorch', '3SsnPorch',
       'ScreenPorch', 'PoolArea', 'MiscVal', 'MoSold', 'YrSold', 'SaleType',
       'SaleCondition', 'SalePrice']
numerical_cols=["LotFrontage","MasVnrArea","GarageYrBlt"]
categorical_cols=["Alley","MasVnrType","BsmtQual","BsmtCond","BsmtExposure","BsmtFinType1","BsmtFinType2","Electrical","FireplaceQu","GarageType",
                  "GarageFinish","GarageQual","GarageCond","PoolQC","Fence","MiscFeature"]
```

```python
def same(x):
    return(x)
```

```python
passthrough=FunctionTransformer(same)
```

```python
num_pipeline=Pipeline([("imputer",SimpleImputer(strategy='median'))])
cat_pipeline=Pipeline([("imputer",SimpleImputer(strategy='most_frequent'))])
```

```python
impute=ColumnTransformer([("pass",passthrough,no_trans),('num',num_pipeline,numerical_cols),('cat',cat_pipeline,categorical_cols)])
```

```python
data=impute.fit_transform(data)
```

```python
data=pd.DataFrame(data,columns=no_trans+numerical_cols+categorical_cols)
data.isnull().sum()
```

```python
nulls=data.isnull().sum()
print(nulls.to_string())
```

```python
data.shape
```

```python
X=data.drop("SalePrice",axis=1)
y=data["SalePrice"]
```

```python
X
```

```python
y
```

```python
for col in X.columns:
    X[col]=pd.to_numeric(X[col],errors="ignore")
```

```python
X.info()
```

```python
year_features=["YearBuilt","YearRemodAdd","YrSold"]
num_features=X.select_dtypes(include=["int64","float64"]).columns.tolist()
cat_features=X.select_dtypes(include=['object','category']).columns.tolist()
```

```python
cat_features
```

```python
num_features=[col for col in num_features if col not in year_features]
num_features
```

```python
preprocessor=ColumnTransformer([("num",StandardScaler(),num_features),("date",passthrough,year_features),("cat",OneHotEncoder(handle_unknown="ignore",sparse_output=False),cat_features)])
```

```python
preprocessor
```

```python
from sklearn.linear_model import Lasso
```

```python
X_transformed=preprocessor.fit_transform(X)
X_transformed
```

```python
X_transformed.shape
```

```python
y.shape
```

```python
feature_names=[]
for name,transformer,cols in preprocessor.transformers_:
    if transformer=='passthrough':
        if isinstance(cols,list):
            feature_names.extend(cols)
        else:
            feature_names.append(cols)
    elif isinstance(transformer,Pipeline):
        last_step=transformer.steps[-1][1]
        if hasattr(last_step,'get_feature_name_out'):
            transformed=last_step.get_feature_names_out(cols)
            feature_names.extend([f"{name}__{f}" for f in transformed])
        else:
            feature_names.extend([f"{name}__{col}"for col in cols])
    elif hasattr(transformer,'get_feature_names_out'):
        transformed=transformer.get_feature_names_out(cols)
        feature_names.extend([f"{name}__{f}" for f in transformed])
    else:
        feature_names.extend([f"{name}__{col}" for col in cols])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_transformed,y,test_size=0.2,random_state=3)
```

```python
from sklearn.linear_model import LassoCV
```

```python
lasso_cv=LassoCV(cv=5,random_state=3)
lasso_cv.fit(X_train,y_train)
print("Best alpha:",lasso_cv.alpha_)
```

```python
lasso=Lasso(alpha=1272,random_state=3)
lasso.fit(X_train,y_train)
```

```python
lasso.coef_.shape
```

```python
selected_mask=lasso.coef_!=0
selected_features=np.array(feature_names)[selected_mask]
selected_features
```

```python
selected_mask.sum()
```

```python
X_train_selected=X_train[:,selected_mask]
X_test_selected=X_test[:,selected_mask]
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train_selected,y_train)
```

```python
y_train_pred_lr=lr.predict(X_train_selected)
y_train_pred_lr[:10]
```

```python
y_train[:10]
```

```python
y_pred_lr=lr.predict(X_test_selected)
y_pred_lr[:10]
```

```python
y_test[:10]
```

```python
r2_lr=r2_score(y_pred_lr,y_test)
mae_lr=mean_absolute_error(y_pred_lr,y_test)
mse_lr=mean_squared_error(y_pred_lr,y_test)
relative_error_lr=mae_lr/(np.mean(y_test))*100
print("r2score lr:",r2_lr)
print("mean absolute error lr:",mae_lr)
print("mean squared error lr:",mse_lr)
print("relative error lr:",relative_error_lr)
```

```python
from sklearn.neighbors import KNeighborsRegressor
train_score=[]
test_score=[]
k_range=(1,20)
for k in k_range:
    knn=KNeighborsRegressor(n_neighbors=k)
    knn.fit(X_train_selected,y_train)
    y_train_pred_knn=knn.predict(X_train_selected)
    y_pred_knn=knn.predict(X_test_selected)
    train_score.append(r2_score(y_train_pred_knn,y_train))
    test_score.append(r2_score(y_pred_knn,y_test))
plt.figure(figsize=(15,10))
plt.plot(k_range,train_score,label='train score',marker="+")
plt.plot(k_range,test_score,label='test score',marker='o')
plt.xlabel("n_neighbors")
plt.ylabel("Score")
plt.title("K vs Score")
plt.grid(True)
plt.legend()
plt.show()
```
